```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
```

```python
file = '../Data_sources/coinbase_historical_pricechange.csv'
df = pd.read_csv(file)
df.head()
```

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'], unit = 's')
df.head()
```

```python
df.set_index(['Timestamp'], inplace = True)
```

```python
df.head()
```

```python
indexed = df.ix['2017-06-01 00:00:00':'2018-06-01 00:00:00']
indexed.head()
```

```python
apr_jun_2018 = df.ix['2018-04-01 00:00:00':'2018-06-01 00:00:00']
apr_jun_2018.head()
```

```python
plt.plot(apr_jun_2018.index.values,apr_jun_2018['Close'])
plt.xlabel('Date')
plt.ylabel('USD')
plt.xticks(rotation = 'vertical')
plt.tight_layout()
plt.show()
plt.savefig('../Output/btc_apr_jun_2018.png')
```
